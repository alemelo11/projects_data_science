### **Q4. Há *outliers* presentes?**

```python
# view the statistical summary of the numerical variables
df[['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month',
    'calculated_host_listings_count', 'availability_365']].describe()
```

#### Definindo valores para `price`

A variável `price` apresenta uma média de 256.21, indicando preços razoáveis para a maioria das acomodações, mas com grande variação devido ao desvio padrão de 214.56. A mediana é 209,00.

Há um preço mínimo de 25,00 e um máximo de 4786,00, evidenciando a presença de acomodações extremamente caras, provavelmente de luxo, ao lado de opções acessíveis.



```python
# Calculating IQR for 'price'
q1_price = df['price'].quantile(0.25)
q3_price = df['price'].quantile(0.75)
iqr_price = q3_price - q1_price

# Defining limits for 'price'
upper_limit_price = q3_price + 1.5 * iqr_price
lower_limit_price = max(0, q1_price - 1.5 * iqr_price)  # Ensure lower limit is not negative

print(f"IQR of the price variable: {iqr_price:.2f}")
print(f"Upper limit for price: {upper_limit_price:.2f}")
print(f"Lower limit for price: {lower_limit_price:.2f}")
```

#### Boxplot para `price`

```python
# price
df.price.plot(kind='box', vert=False, figsize=(15, 3))
plt.show()

# check the number of values above 544.5 for price
print('\nPrice: values above 581.5')
print(f'{len(df[df.price > 581.5])} entries')
print(f'{(len(df[df.price > 581.5]) / df.shape[0]) * 100:.4f}%')
```



#### Definindo limites e Boxplot para `minimum_nights`

A variável `minimum_nights`, com uma mediana de 2 noites, revela que, em sua maior parte, as propriedades tendem a exigir estadias curtas. No entanto, a presença de um valor máximo de 250 noites indica que algumas propriedades possuem requisitos excepcionais de permanência. O IQR (intervalo interquartílico) de 3 noites e o limite superior de 8,5 noites sugerem que a maioria das propriedades se concentra em estadias mais curtas, com algumas exceções.

```python
# Calculating limits for 'minimum_nights'
q1_min_nights = df['minimum_nights'].quantile(0.25)
q3_min_nights = df['minimum_nights'].quantile(0.75)
iqr_min_nights = q3_min_nights - q1_min_nights
inf_min_nights = max(0, q1_min_nights - 1.5 * iqr_min_nights)  # Lower limit cannot be negative
sup_min_nights = q3_min_nights + 1.5 * iqr_min_nights

# Output for minimum_nights
print("IQR of the minimum_nights variable: ", round(iqr_min_nights, 2))
print("Upper limit for minimum_nights: ", round(sup_min_nights, 2))
print("Lower limit for minimum_nights: ", round(inf_min_nights, 2))


# Plotting the 'minimum_nights' boxplot
df.minimum_nights.plot(kind='box', vert=False, figsize=(15, 3))
plt.title('Boxplot of Minimum Nights')
plt.xlabel('Minimum Nights')
plt.show()

# Check the number of values above the upper limit for 'minimum_nights'
values_above_upper_limit = len(df[df.minimum_nights > sup_min_nights])

# Total entries in the dataset
total_entries = df.shape[0]

# Output the number of outliers and percentage
print(f"minimum_nights: values above: {sup_min_nights:.1f}:")
print(f'minimum_nights: values above: {values_above_upper_limit} entries')
print(f"minimum_nights: values above {(values_above_upper_limit / total_entries) * 100:.4f}%")


# see the average and median of the `minimum_nights' column``
print(f"minimum_nights mean: {df.minimum_nights.mean():.2f}")
print(f'minimum_nights median:{df.minimum_nights.median()}')




```

### **Q6. Qual a correlação existente entre as variáveis**

Correlação significa que existe uma relação entre duas coisas. No nosso contexto, estamos buscando relação ou semelhança entre duas variáveis.

Essa relação pode ser medida, e é função do coeficiente de correlação estabelecer qual a intensidade dela. Para identificar as correlações existentes entre as variáveis de interesse, vou:

* Criar uma matriz de correlação
* Gerar um *heatmap* a partir dessa matriz, usando a biblioteca `seaborn`

```python
# create a correlation matrix
corr = df_clean[['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month',
    'calculated_host_listings_count', 'availability_365']].corr()
# show the correlation matrix
display(corr)
```

```python
sns.heatmap(corr, cmap='RdBu', fmt='.2f', square=True, linecolor='white', annot=True);
```

É interessante notar que a correlação entre`price` e `availability_365` (apenas 0,09) é bem baixa. Esperava-se que propriedades mais caras tivessem menos disponibilidade, já que a demanda costuma ser menor. Isso pode sugerir que o mercado está segmentado, como no caso de aluguéis de luxo em comparação com os econômicos.
